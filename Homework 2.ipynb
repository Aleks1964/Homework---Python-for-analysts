#### 5.1 Выберите дом клиенту

Клиент хочет дом с видом на набережную, как минимум с тремя ванными и с подвалом. Сколько вариантов есть у клиента?

_Дома с как минимум с тремя ванными:_

```python
df[df['bathrooms'] >= 3]
```

_Дома с подвалами:_

```python
df[df['sqft_basement'] > 0]
```

_Дома с видом на набережную:_

```python
df[df['waterfront'] != 0]
```

```python
f"Количество домов с видом на набережную, как минимум с тремя ванными и с подвалом равно {df[(df['waterfront'] != 0) & (df['bathrooms'] >= 3) & (df['sqft_basement'] > 0)].shape[0]}."
```

_Варианты искомых домов:_

```python
df[(df['waterfront'] != 0) & (df['bathrooms'] >= 3) & (df['sqft_basement'] > 0)]
```

#### 5.2 Выберите дом клиенту

Клиент хочет дом либо с очень красивым видом из окна, либо с видом на набережную, в очень хорошем состоянии и год постройки не меньше 1980 года. В какой ценовом диапазоне будут дома?

_Дома с очень красивым видом из окна:_

```python
df[df['view'] == 4]
```

_Дома с с видом на набережную:_

```python
df[df['waterfront'] == 1]
```

_Дома в очень хорошем состоянии:_

```python
df[df['condition'] == 5]
```

_Дома, год постройки которых не ранее 1980 года:_

```python
df[df['yr_built'] >= 1980]
```

 _Варианты искомых домов:_

```python
df[((df['view'] == 4) | (df['waterfront'] == 1)) & (df['condition'] == 5) & (df['yr_built'] >= 1980)]
```

_Цены искомых домов:_

```python
df[((df['view'] == 4) | (df['waterfront'] == 1)) & (df['condition'] == 5) & (df['yr_built'] >= 1980)]['price']
```

_Ценовой диапазон искомых домов:_

```python
looking_for_houses = df[((df['view'] == 4) | (df['waterfront'] == 1)) & (df['condition'] == 5) & (df['yr_built'] >= 1980)]['price']
looking_for_houses.min(), looking_for_houses.max()
```
